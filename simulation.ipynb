```python
picks = [(i*10)+3 if i%2==0 else (i*10)+7 for i in range(16)]

def getPotentialPicks(picks,draft):
    draft['low']=draft['med_pick']-draft['standard_deviation']
    draft['high'] = draft['med_pick']+draft['standard_deviation']
    for pick in picks:
        print(f'with the {pick} pick you can probably select: ')
        print(draft[['PLAYER','med_pick','low','high']].loc[(draft['low']<=pick) & (pick<=draft['high']) & (draft['number_of_drafts']>=500)])
getPotentialPicks(picks,mockDraft)
```
