```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import copy
```

```python
def getFantasyPerformance(link):
    response = requests.get(link)
    html=response.text.encode()
    soup = BeautifulSoup(html,'html.parser')
    table = soup.find('div',{'id':'div_fantasy'}).find('tbody')
    players = [x.find_all('td') for x in table.find_all('tr')]
    players = [[y.text.replace('*','').replace('+','') for y in x] for x in players]
    return players

def iterateYears(start_year=2010,end_year=2018):
    columns = ['PLAYER','TEAM','POS','AGE','GP','GS','Passing_Cmp','Passing_Att','Passing_Yds','Passing_TD','Int',
          'Rushing_Att','Rushing_Yds','Y/A','Rushing_TD','Tgt','Rec','Reception_Yds','Y/R','Rec_TD','2PM','2PP',
          'FantPt','PPR','DKPt','FDPt','VBD','PosRank','OvRank']
    data = {year:pd.DataFrame(data=getFantasyPerformance(f'https://www.pro-football-reference.com/years/{year}/fantasy.htm'),columns=columns) for year in range(start_year,end_year+1)}
    return data

def changeDataType(data):
    dataCopy=copy.deepcopy(data)
    for year in dataCopy:
        dataCopy[year] = dataCopy[year].replace(r'', np.nan, regex=True)
        for col in dataCopy[year].columns:
            if col not in ['PLAYER','TEAM','POS']:
                dataCopy[year][col]=dataCopy[year][col].astype(float)
    return dataCopy
        
       
```

```python
fantasyPtsYear = iterateYears()
fantasyPtsYear = changeDataType(fantasyPtsYear)
```

```python
for year in fantasyPtsYear:
    fantasyPtsYear[year].to_csv(f'HistoricalFantasyData/{year}_fantasyStats.csv',index=False)
    
```
