In this script we are going to attempt to predict games based on certain summary statistics. I ideally want to see how game plans unfold for each time and eventually build a dataset that will work towards building these statistics. In football we qualitatively know that certain factors are really good for your teams performance on offense they are as such.

* Offensive Line Performance -- When you have a good O line you have a good base for your offense. We need to figure out a way to measure offensive line performance using the data we've been given

* QB as a passer -- The NFL is a passing league now and quarterbacks are the head of that. What has been used in the past is passer rating, so we will attempt to include that in our dataset.

* Rushing Proliferation -- While it's obvious that good passing plays can help keep your plays alives, Rushing plays a part in that too.

* Third Down Conversion -- How many third downs were converted during that game

* Number of First Downs

Offensive Line performance can be measured generally as number of qb rushes by a defender and/or sacks given up. While certain teams can have qbs that simply hold the ball for too long, we can overall say sacks given up are pretty vital.

QB as a passer is measured by the passer rating

Rushing Proliferation will be percentage of Rushes that go for more than >= 5yds and < 5yds

On Defense you also have certain metrics which we can include in the model.

* Interception Pct -- The amount of passes that result in an interception

* Rush Fumble Pct -- The amount of rushes that result in a fumble

* Pass Fumble Pct -- The amount of passes that result in a fumble

* Sack Pct -- The pct of sacks by the defense

* Pick Six Pct

* Red Zone Efficiency

Special Teams

## Notes

* We can build a thing of simulations where we have P(pass) and P(rush) and which way those go based on the previous result and then if it'll be a sack, fumble, etc.

* We can use clustering to find similar match-ups

```python
def buildGameSummary(data):
    '''
    Using the play by play data, we are going to attempt to build a dataset to describe football games.
    '''
    #print(list(data.columns))
    buildData = data[[x for x in data.columns if 'Unnamed' not in x]].sort_values(['GameId','Quarter','Minute','Second'])
    buildData['PlayCount']=1
    buildDataSacks = offensiveLinePerformance(buildData)
    buildDataPasser = passerRating(buildData)
    buildDataRush = rushingProliferation(buildData)
    buildDataInt = intLikelihood(buildData)
    buildDataFumble = fumbleLikelihood(buildData)
    buildDataThirdDown = thirdDownConversion(buildData)
    buildDataTeamWin = teamWin(buildData)
    dataset = buildDataPasser.merge(buildDataRush,how='left',on=['GameId','OffenseTeam']).merge(buildDataSacks,how='left',on=['GameId','OffenseTeam']).merge(buildDataInt,how='left',left_on=['OffenseTeam','GameId'],right_on=['DefenseTeam','GameId']).merge(buildDataFumble,how='left',left_on=['OffenseTeam','GameId'],right_on=['DefenseTeam','GameId']).merge(buildDataThirdDown,how='left',on=['OffenseTeam','GameId']).merge(buildDataTeamWin,how='left',left_on=['GameId','OffenseTeam'],right_on=['GameId','WhichTeamWin'])
    dataset['TeamWin']=dataset['TeamWin'].fillna(0)
    
    dataset = dataset.drop(['WhichTeamWin','DefenseTeam_x','DefenseTeam_y'],axis=1).rename(columns={'OffenseTeam':'Team'})
    return dataset

def offensiveLinePerformance(data):
    dataSacks = data.groupby(['GameId','OffenseTeam']).sum().reset_index()[['GameId','OffenseTeam','IsSack','PlayCount']]
    dataSacks['SackPct'] = dataSacks['IsSack']/dataSacks['PlayCount']
    return dataSacks[['GameId','OffenseTeam','SackPct']]

def passerRating(data):
    '''
    Use passing downs to get each teams overall passer rating. The formula is as follows:
    
    a = (COMP/ATT - .3) * 5
    b = (YDS/ATT - 3) * .25
    c = (TD/ATT) * 20
    d = 2.375 - (INT/ATT * 25)
    
    if any result is >2.375 then set it to 2.375 if its negative set it to 0
    '''
    
    passerData = data.loc[data['IsPass']==1]
    passerData['Att'] = 1
    passerData = passerData.groupby(['GameId','OffenseTeam']).sum().reset_index()[['GameId','OffenseTeam','Att','IsIncomplete','Yards','IsTouchdown','IsInterception']]
    passerData['IsComplete'] = passerData['Att'] - passerData['IsIncomplete']
    passerData['a'] = 5*((passerData['IsComplete']/passerData['Att']) - .3)
    passerData['b'] = .25 * ((passerData['Yards']/passerData['Att']) - 3)
    passerData['c'] = 20 * passerData['IsTouchdown']/passerData['Att']
    passerData['d'] = 2.375 - ((passerData['IsInterception']/passerData['Att']) *25)
    passerData['a'] = passerRatingCheck(passerData['a'])
    passerData['b'] = passerRatingCheck(passerData['b'])
    passerData['c'] = passerRatingCheck(passerData['c'])
    passerData['d'] = passerRatingCheck(passerData['d'])
    passerData['PasserRating'] = ((passerData['a']+passerData['b']+passerData['c']+passerData['d'])/6) * 100
    return passerData[['GameId','OffenseTeam','PasserRating']]

def passerRatingCheck(series):
    
    '''
    If any result is >2.375 then set it to 2.375 if its negative set it to 0
    '''
    
    editedSeries = [2.375 if x>2.375 else 0 if x<0 else x for x in series]
    return series

def rushingProliferation(data):
    '''
    Rushing proliferation in this context means how the run game can keep a drive alive.
    We define this by all plays that go for greater than 5 yards.
    '''
    rushData = data.loc[data['IsRush']==1][['GameId','OffenseTeam','IsRush','Yards']]
    rushData['<5'] = [1 if x<5 else 0 for x in rushData['Yards']]
    rushData['>=5'] = [1 if x>=5 else 0 for x in rushData['Yards']]
    rushData = rushData.groupby(['GameId','OffenseTeam',]).sum().reset_index()[['GameId','OffenseTeam','IsRush','<5','>=5']]
    rushData['smallGainPct'] = rushData['<5']/rushData['IsRush']
    rushData['bigGainPct'] = rushData['>=5']/rushData['IsRush']
    rushData['gainRatio'] = rushData['>=5']/rushData['<5']
    return rushData[['GameId','OffenseTeam','bigGainPct']]

def intLikelihood(data):
    '''
    Calculate the percentage of interceptions that occurred in a given game.
    '''
    intData = data.groupby(['GameId','DefenseTeam']).sum().reset_index()[['GameId','DefenseTeam','IsInterception','IsPass']]
    intData['IntPct'] = intData['IsInterception']/intData['IsPass']
    return intData[['GameId','DefenseTeam','IntPct']]

def fumbleLikelihood(data):
    '''
    Calculate the percentage of interceptions that ocurred in a given game.
    '''
    
    fumbleData = data[['GameId','DefenseTeam','IsFumble','PlayCount']].groupby(['GameId','DefenseTeam']).sum().reset_index()
    fumbleData['FumblePct'] = fumbleData['IsFumble']/fumbleData['PlayCount']
    
    return fumbleData[['GameId','DefenseTeam','FumblePct']]

def thirdDownConversion(data):
    '''
    Take the total number of third downs for each team and see the conversion 
    '''
    
    thirdDownConv = data.loc[data['Down']==3].groupby(['GameId','OffenseTeam']).sum().reset_index()
    thirdDownConv['ThirdDownConversion'] = thirdDownConv['SeriesFirstDown']/thirdDownConv['PlayCount']
    return thirdDownConv[['GameId','OffenseTeam','ThirdDownConversion']]

def numOffenseTouchdown(data):
    '''
    Get the total number of Offense Touchdowns 
    '''
    
    offenseTD = data.loc[(data['IsFumble']!=1) & (data['IsInterception']!=1)]
    offenseTD = offenseTD.groupby(['GameId','OffenseTeam']).sum().reset_index()
    return offenseTD[['GameId','OffenseTeam','IsTouchdown']]

def numDefenseTouchdown(data):
    '''
    Get the total number of Defense Touchdowns
    '''
    
    defenseTD = data.loc[((data['IsFumble']==1) | (data['IsInterception']==1))]
    defenseTD = defenseTD.groupby(['GameId','DefenseTeam']).sum().reset_index()
    return defenseTD[['GameId','DefenseTeam','IsTouchdown']]

def totalTouchdown(data):

    offenseTD = numOffenseTouchdown(data)
    defenseTD = numDefenseTouchdown(data)
    
    totTD = offenseTD.merge(defenseTD,how='outer',left_on=['GameId','OffenseTeam'],right_on=['GameId','DefenseTeam'])
    totTD['OffenseTeam']=totTD['OffenseTeam'].fillna(totTD['DefenseTeam'])
    totTD['IsTouchdown_x'] = totTD['IsTouchdown_x'].fillna(0)
    totTD['IsTouchdown_y'] = 0
    totTD['NumTouchdowns'] = totTD['IsTouchdown_x'] + totTD['IsTouchdown_y']
    return totTD[['GameId','OffenseTeam','NumTouchdowns']].pivot(index='GameId',columns='OffenseTeam',values='NumTouchdowns')

def teamWin(data):
    totTD = totalTouchdown(data)
    teamWin = totTD.idxmax(axis=1).reset_index().rename(columns={0:'WhichTeamWin'})
    teamWin['TeamWin']=1
    return teamWin

def normalize(data):
    scaler = MinMaxScaler()
    prenorm = data.drop(['GameId','Team','TeamWin'],axis=1)
    X_norm = pd.DataFrame(columns=list(prenorm.columns),data=scaler.fit_transform(prenorm))
    y_data = data['TeamWin']
    return X_norm, y_data

def feature_importance(coefficients):
    zScore = (coefficients - np.mean(coefficients))/np.std(coefficients,ddof=1)
    return zScore

def sc_evaluate_clusters(X,max_clusters):
    s = np.zeros(max_clusters+1)
    s[0] = 0;
    s[1] = 0;
    for k in range(2,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        s[k] = silhouette_score(X,kmeans.labels_,metric='euclidean')
    plt.plot(range(2,len(s)),s[2:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Silhouette Score')
```

```python
combinedData = pd.concat([buildGameSummary(pbp2014),buildGameSummary(pbp2015),buildGameSummary(pbp2016),buildGameSummary(pbp2017)]).reset_index(drop=True)
X_norm, y = normalize(combinedData)
```

```python
X_norm.corr()
```

Some intersting notes

* A teams ability to keep a drive a live is pretty valuable to a passer rating. Expected behavior.

* Getting sacked is negatively correlated with third down conversions. Also expected.

* There is some what a positive correlation (albeit weak) between the number of interceptions you get and passer rating. How I can rationalize it is the defense gives you more chances to get back on the field.

```python
np.random.seed(146)
X_train, X_test, y_train, y_test = train_test_split(X_norm, y,test_size = 0.33)
cRand = np.random.rand(1000)
rs = RandomizedSearchCV(estimator=LogisticRegression(),param_distributions={'C':cRand},n_iter=100,scoring='roc_auc')
rs.fit(X_train,y_train)
bestC = rs.best_params_['C']
model = LogisticRegression(C=bestC)
model.fit(X_train,y_train)
y_proba = model.predict_proba(X_test)[:,1]

print(roc_auc_score(y_score=y_proba,y_true=y_test))
importances = pd.DataFrame(index=X_train.columns,data=model.coef_.T)
importances.plot(kind='barh')
```

## Findings of Model

* Big gains by RBs are not that valuable for winning games. 
* Passer rating has the highest coefficient by far
* Removing that we see that sack percentage is valuage followed by ints, fumbles and then third down conversions.
