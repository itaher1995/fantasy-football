```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

```python
LINK = 'https://overthecap.com/salary-cap-space'
YEAR = 2018
```

```python
response = requests.get(LINK)
html = response.text.encode()
soup = BeautifulSoup(html,'html.parser')
```

```python
body = [[re.sub(r'((?:49ers|Lions|Raiders|Ravens|Colts|Saints|Vikings|Seahawks|Steelers|Giants|Chargers|Patriots|Panthers|Packers|Chiefs|Redskins|Buccanners|Falcons|Cowboys|Jaguars|Redskins|Cardinals|Browns|Bengals|Broncos|Chargers|Titans|Rams|Texans|Chiefs|Bears|Eagles|Dolpins))(\$\d+[0-9,]+)\$(.*)',r'\1\2',y.text).replace(',','').split('$')[:2] for y in x.find('tbody').find_all('tr')] for x in soup.find_all('table',{'class':'sortable salary-cap-space-table'})]
tables=[pd.DataFrame(columns=['Team','Cap_Space'],data=table) for table in body]

year = 2015
for t in tables:
    t['Year']=year
    year+=1
capSpace = pd.concat(tables)
capSpace = capSpace.loc[capSpace['Year']<=YEAR]
capSpace['Cap_Space'] = capSpace['Cap_Space'].apply(lambda x: x.replace(')(',''))
capSpace['Cap_Space']=capSpace['Cap_Space'].astype(int)
capSpace
# table = soup.find('tbody')
# [[y.text for y in x.find_all('a')]+[int([y.text for y in x.find_all('td')][2][1:].replace(',',''))] for x in table.find_all('tr')]
# #[re.sub(r'([A-z]+ [A-Z][a-z]+)[ ]?((?:49ers|Lions|Raiders|Ravens|Colts|Saints|Vikings|Seahawks|Steelers|Giants|Chargers|Patriots|Panthers|Packers|Chiefs|Redskins|Buccanners|Falcons|Cowboys|Jaguars|Redskins|Cardinals|Browns|Bengals|Broncos|Chargers|Titans|Rams|Texans|Chiefs|Bears|Eagles))(\$.*\(\$)',r'\1,\2,\3',x.text) for x in table.find_all('tr')]
```

```python
capSpace.to_csv('HistoricalRegSeasonData/cap_space.csv')
```
