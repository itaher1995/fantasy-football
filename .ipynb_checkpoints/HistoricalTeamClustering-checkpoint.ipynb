```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('HistoricalFantasyData/historicalFantasyTeamsClusterData.csv')
data
```

```python
def sc_evaluate_clusters(X,max_clusters):
    s = np.zeros(max_clusters+1)
    s[0] = 0;
    s[1] = 0;
    for k in range(2,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        s[k] = silhouette_score(X,kmeans.labels_,metric='euclidean')
    plt.plot(range(2,len(s)),s[2:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Silhouette Score')
```

## Looking at clusters using position results

Here we want to understand how positions effect the clustering result and maybe we can identify types of teams that belong to each cluster.

```python
sc_evaluate_clusters(data.drop('TEAM_YEAR',axis=1),10)
```

```python
np.random.seed(101)
kmeans = KMeans(init='k-means++', n_clusters=7, n_init=10)
data['kmeans'] = kmeans.fit_predict(data.drop('TEAM_YEAR',axis=1))
data
```

```python
for i in range(7):
    print(i)
    print(data.loc[data['kmeans']==i])
```

```python
centers = pd.DataFrame(kmeans.cluster_centers_,columns=list(data.columns)[1:-1])
centers['tot'] = np.sum(centers,axis=1)
centers
```

```python
centers.sort_values('tot')
```

```python
data['TEAM']=data['TEAM_YEAR'].str.slice(start=0,stop=3)
teamProg = {key:list(data['kmeans'].iloc[data.groupby('TEAM').groups[key]]) for key in data.groupby('TEAM').groups}
teamProg 
```

```python
{key:['top 3' if x in [3,4,0] else 'middle of road' if x==5 else 'bottom 3' for x in teamProg[key]] for key in teamProg}
```

Notes

* Cluster 3 outperforms the rest. We see teams with great RB and WR1 performances (think Steelers and last years Saints.)

* Cluster 4 is where New England has found itself over the last 8 years. We see high performances by Quarterbacks and Tight Ends in this group. In particular we see weak performances by running backs (New England was never known for its backs.)

* Cluster 0 is a very Lions for the last couple years scenario. Strong QB and Strong Wide Receivers.

* Cluster 5 is the middle of the road for the fantasy rosters. Average QB, strong RB (2017 Rams show up in this cluster.) and everything else is lackluster.

* Cluster 1's only saving grace is its somewhat weak QB performance. WR1 is also ok,but not great. Dallas lasy year and Cincinatti last year also show up in this cluster.

* Cluster 6 has a strong wide receiver showing. Arizona is categorized by this, probably due to Larry Fitzgerald.

* Cluster 2 is pretty much your worst teams. We see Cleveland, the Rams for the most part and Chicago for the last two years show up.

We see that teams that have been at the bottom, tend to stay at the bottom and teams that are the top tend to stay at the top. Sometimes some teams go to middle and then up or middle and then down. The only team with a weird pattern is IND which is all over the place.
